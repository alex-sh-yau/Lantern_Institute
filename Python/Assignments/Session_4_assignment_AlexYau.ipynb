#### Titanic Insights
There are quite a few insights that can be seen from analysing the Titanic's passenger data. From a data science perspective, if we were to look at the survivability of passengers and what variables affect them, some conclusions can be made:

   - The Cabin column is correlated to Pclass so it can be dropped.
    
   - Fare is also correlated to Pclass, however one is continuous and can be used for regression analysis while the other is categorical and can be used for classification analysis.
   - SibSp and Parch are correlated to Ticket and can be dropped.
   - For predicting survivability, Ticket does not seem to give any insights so it can be dropped.
   - Name was useful for discovering meaning of other columns, but for anaysis of survivability, can be dropped.
   - Sex and Embarked can be useful for classification analysis. Age can be used for regression analysis to a certain degree

## Sleep Data - Part 1: Warm-up

```python
import matplotlib
import matplotlib.pyplot as plt
```

#### Bedtime Hours

```python
sleep_df = pd.read_csv('sleepdata.csv', sep=';')
```

```python
sleep_df.head()
```

```python
sleep_df['Start'] = pd.to_datetime(sleep_df['Start'])
sleep_df['End'] = pd.to_datetime(sleep_df['End'])
```

```python
start_df = (sleep_df['Start'].dt.hour + sleep_df['Start'].dt.minute/60 + sleep_df['Start'].dt.second/3600)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('Hour of Day at Start of Bedtime')
ax.set_ylabel('Frequency')
ax.hist(start_df, bins=24)
```

#### Conclusion
This individual sleeps very late, always after 12AM, and most often around 2-4AM. 

#### Sleep Duration

```python
start_df[start_df>20] = start_df[start_df>20] - 24 # Convert times before 12AM for numerical calculation
end_df = (sleep_df['End'].dt.hour + sleep_df['End'].dt.minute/60 + sleep_df['End'].dt.second/3600)

duration_df = (end_df - start_df) * 60

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('Sleep Time (min)')
ax.set_ylabel('Frequency')
ax.hist(duration_df, bins=50)
```

#### Conclusion
The amount of time this person sleeps for is approximated by a normal distribution. The median is around 360 minutes which is 6 hours, and at the extremes are some nights with less than 2 hours of sleep and almost 12 hours of sleep.

#### Sleep Duration Over Time:

```python
fig = plt.figure(figsize=(20, 6))
ax = fig.add_subplot(111)
ax.set_xlabel('Date')
ax.set_ylabel('Sleep Time (min)')
ax.scatter(sleep_df['End'].dt.date, duration_df)
```

#### Conclusion
The majority of the data starts around December 2012 and is quite consistent over time, but there is some data missing around August 2013 where the person may not have recorded their sleep.

Looking at the scatterplot above, generally the duration of sleep has slightly increased over time.

#### Sleep Quality Over Time - Does Duration Mean Quality?

```python
quality_df = sleep_df['Sleep quality'].str[:-1].astype(int)
```

```python
fig = plt.figure(figsize=(20, 6))
ax = fig.add_subplot(111)
ax.set_xlabel('Date')
ax.set_ylabel('Sleep Time (min)')
ax.scatter(sleep_df['End'].dt.date, duration_df, c=quality_df, label=quality_df, cmap=plt.cm.autumn)

# Quality is added in here to the scatter point colors
# Yellow means higher quality, red means lower quality
```

```python
fig = plt.figure(figsize=(20, 6))
ax = fig.add_subplot(111)
ax.set_xlabel('Sleep Quality (%)')
ax.set_ylabel('Sleep Time (min)')
ax.scatter(quality_df, duration_df)
```

#### Conclusion
Adding in sleep quality as a dimension when looking at sleep duration over time, it shows very clearly that more sleep generally equates to higher sleep quality (yellow scatter points) and less sleep equates to lower sleep quality (red scatter points). This is further illustrated when looking at just sleep quality versus sleep duration on a scatter plot, which clearly shows a linear correlation.

There are definitely some outliers with low sleep quality on nights with more than 8 hours of sleep.
