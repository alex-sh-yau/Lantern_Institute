### Gen 1 moves :

```python
# Write a query to output the name of the move and name of the type for Generation 1 Pokemon. 
#You wanna look into move_names,moves and type_names tables
cur.execute("""SELECT m.identifier, t.name 
            FROM moves m INNER JOIN type_names t 
            ON m.type_id = t.type_id 
            WHERE m.generation_id IS 1 AND t.local_language_id IS 9 
            ORDER BY t.type_id;""")
cur.fetchall()
```

### It is hard to write custome queries each time. Lets use the power of Pandas/Python by getting everything into Pandas DF:

```python
# Use "for loops" to read each table in the DB and store it as a df with the same name as the table.
#You should see a set of Pandas df like this :

df_p = {}
pokemon_data = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", con)
for i in pokemon_data.index:
    df_p['{}_df'.format(pokemon_data.iloc[i,0])] = pd.read_sql_query("SELECT * FROM {};".format(pokemon_data.iloc[i,0]), con)
for key, value in df_p.items():
    print(key)
```

```python
# It seems hard to see the correlation based on the scatter plots.

# Instead, use "pearsonr" function from scipy's "stats" library to compute the correlation between 
from scipy import stats
stats.pearsonr?
```

```python
# Loop over all combinations and output the correlation coefficients. Then sort based on the coefficients
for (columnName_x, columnData_x) in poke_statistics.iteritems():
    for (columnName_y, columnData_y) in poke_statistics.iteritems():
        print(str(columnName_x) + ' and ' + str(columnName_y) + ':', end = '') 
        print(stats.pearsonr(columnData_x, columnData_y))
```

```python
# Nothing is better than a good visualization. 
# Try seaborn's pairplot to get something like this :
sns.pairplot(poke_statistics, kind='reg')
```

```python
# Get rid of extra columns :
poke_statistics = poke_statistics[['hp','attack','defense','sp. atk', 'sp. def', 'speed']]
sns.pairplot(poke_statistics[['hp','attack','defense','sp. atk', 'sp. def', 'speed']], kind='reg')
```

```python
# See if you can refine your code to get this :

def corrfunc(x, y, **kws):
    (r, p) = stats.pearsonr(x, y)
    ax = plt.gca()
    ax.annotate("r = {:.2f} ".format(r),
                xy=(.7, .9), xycoords=ax.transAxes)

ax = sns.pairplot(poke_statistics, kind='reg')
ax.map(corrfunc)
```

#### Summarize your observations/conclusions here :

In general, most of the Pokemon statistical attributes are slightly positively correlated to each other according to their Pearson correlation coefficients. The exception to this is Speed-Defense, which are almost completely uncorrelated to each other. 

The next stat pairs with the lowest positive correlation to each other are Speed-HP, Attack-Sp.Def, and Speed-Sp.Def.

The stat pairs with the highest positive correlation to each other are Defense-Sp.Def and Sp.Atk-Sp.Def. This means that in general, Pokemon with high Sp.Def stats also have high Defense and Sp.Atk stats.

### What are the strongest and weakest Pokemon species?

```python
# Do your calculations here :
pokedex_total = pd.concat([pokedex, poke_statistics.sum(axis=1)], axis=1)
single_types_total = pokedex_total[pokedex_total['type 2'] == '']
dual_types_total = pokedex_total[pokedex_total['type 2'] != '']
```

```python
fig = plt.figure(figsize=(10,10))
ax = sns.barplot(single_types_total.groupby(['type 1'])[0].mean().index, single_types_total.groupby(['type 1'])[0].mean())
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Average Strength of Single Type Pokemon')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax = sns.heatmap(dual_types_total.groupby(['type 1', 'type 2'])[0].mean().unstack(), cmap=sns.cubehelix_palette(8), 
                 annot=True, linewidth=0.5, fmt='.0f')
ax.set_title('Average Strength of Dual Type Pokemon')
plt.show()
```

#### Explain why using "Total" is not the best metric ?


The total statistic does not reflect the relative strength of each statistic.

One Pokemon may have very high attack but very low defense and have a better total statistic compared to another Pokemon   with a more even stat distributions. This does not necessarily mean that particular Pokemon is stronger.

```python
# Instead, try a different metric: standardize the six statistic columns independently by converting each 
# value into a z-score so when we do take the sum, we account for the variation in the each statistic using 
# its mean and standard deviation across all Pokemon species. 

poke_stats_z = pd.DataFrame()
for (columnName, columnData) in poke_statistics.iteritems():
    poke_stats_z[columnName] = (poke_statistics[columnName] - poke_statistics[columnName].mean()) / poke_statistics[columnName].std()
poke_stats_z
```

```python
# Define a new column, strength, as the sum of the z-scores of each statisticthe higher this value, 
# the stronger the Pokemon.
poke_stats_z['strength'] = poke_stats_z.sum(axis=1)
poke_stats_z
```

```python
# Repeat the heatmap you created but now for the new strength metric you computed

# I think you mean pairplots?
ax = sns.pairplot(poke_stats_z, kind='reg')
ax.map(corrfunc)
```

### Do any types of Pokemon excel at certain statistics over others?

```python
# Do your calculations here. Create a heat map to show this at the end.

pokedex_stat = pd.concat([pokedex, poke_stats_z['strength']], axis=1)
single_types_stat = pokedex_stat[pokedex_stat['type 2'] == '']
dual_types_stat = pokedex_stat[pokedex_stat['type 2'] != '']
```

```python
fig = plt.figure(figsize=(10,10))
ax = sns.barplot(single_types_stat.groupby(['type 1'])['strength'].mean().index, single_types_stat.groupby(['type 1'])['strength'].mean())
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Average Strength of Single Type Pokemon')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax = sns.heatmap(dual_types_stat.groupby(['type 1', 'type 2'])['strength'].mean().unstack(), cmap=sns.cubehelix_palette(8), 
                 annot=True, linewidth=0.5)
ax.set_title('Average Strength of Dual Type Pokemon')
plt.show()
```
