```python
# Write an SQL statement that returns exactly one row with columns:

# region_id, region_name, province_name, province_code, country_name, country_code for any given regionid.
```

```python
sql_command = """
SELECT
  regionlist.regionid,
  regionlist.regionname,
  regionlist.regionnamelong,
  provinces.province_name,
  provinces.province_code,
  provinces.country_name,
  provinces.country_code
FROM regionlist
JOIN provinces ON regionlist.regionnamelong LIKE '%' || provinces.province_name || '%'
"""
```

```python
pd.read_sql_query(sql_command, con)
```
