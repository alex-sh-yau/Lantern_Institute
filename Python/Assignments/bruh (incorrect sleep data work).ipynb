It seems that the only common elements in the sleep and exercise data are the start and end times. 

However, as seen above, there are no overlapping dates.

Sleep_df starts in 2011 and ends in 2015. Exercise_df starts in 2016 and ends in 2018. 

Therefore to merge these two dataframes for further analysis, an outer join is necessary to keep all data.

Start time and end time of both sleep and exercise together can be used to analyse the person's daily patterns.

```python
sleep_df['Start'] = pd.to_datetime(sleep_df['Start'])
sleep_df['End'] = pd.to_datetime(sleep_df['End'])
```

```python
sleep_df['Start Date'] = sleep_df['Start'].dt.date
sleep_df['Start Time'] = sleep_df['Start'].dt.time
sleep_df['End Time'] = sleep_df['End'].dt.time
```

```python
sleep_df.head()
```

```python
exercise_df['start_time'] = pd.to_datetime(exercise_df['start_time'])
exercise_df['end_time'] = pd.to_datetime(exercise_df['end_time'])
```

```python
# One more thing to deal with first: UTC time offset.
exercise_df.groupby('time_offset').count()
# Since there are no timezones in the sleep data, let us assume for now that the person primarily lives in UTC-0300.
# We shall covert the times accordingly to the UTC-0300 timezone.
```

```python
exercise_df.loc[exercise_df.time_offset == 'UTC-0400', 'start_time'] = exercise_df[exercise_df['time_offset'].isin(['UTC-0400'])]['start_time'] - timedelta(hours=1)
exercise_df.loc[exercise_df.time_offset == 'UTC-0600', 'start_time'] = exercise_df[exercise_df['time_offset'] == 'UTC-0600']['start_time'] - timedelta(hours=3)
exercise_df.loc[exercise_df.time_offset == 'UTC-0700', 'start_time'] = exercise_df[exercise_df['time_offset'] == 'UTC-0700']['start_time'] - timedelta(hours=4)
```

```python
exercise_df.loc[exercise_df.time_offset == 'UTC-0400', 'end_time'] = exercise_df[exercise_df['time_offset'].isin(['UTC-0400'])]['end_time'] - timedelta(hours=1)
exercise_df.loc[exercise_df.time_offset == 'UTC-0600', 'end_time'] = exercise_df[exercise_df['time_offset'] == 'UTC-0600']['end_time'] - timedelta(hours=3)
exercise_df.loc[exercise_df.time_offset == 'UTC-0700', 'end_time'] = exercise_df[exercise_df['time_offset'] == 'UTC-0700']['end_time'] - timedelta(hours=4)
```

```python
exercise_df['Start Date'] = exercise_df['start_time'].dt.date
exercise_df['Start Time'] = exercise_df['start_time'].dt.time
exercise_df['End Time'] = exercise_df['end_time'].dt.time
```

```python
merged_outer = pd.merge(sleep_df, exercise_df, how='outer', on=['Start Date','Start Time','End Time'])
merged_outer
```
